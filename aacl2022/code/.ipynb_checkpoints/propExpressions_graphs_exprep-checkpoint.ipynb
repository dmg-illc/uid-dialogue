```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('proportions.csv')
print(list(df))
```

```python
d_len = df.groupby('dial_id')['uttID'].max()
ids = df.dial_id.unique()
```

```python
d_len.get('S24A')
```

```python
# return proportion through dial we are
def getProp(row):
    res = row.uttID / d_len.get(row.dial_id)
    return res*100

df['percent_dial'] = df.apply(lambda row: getProp(row), axis=1)
    
```

```python
def make_bins(row):
    return int(row.percent_dial) # /5 /10
df['percent_bin'] = df.apply(lambda row: make_bins(row), axis=1)
```

```python
sns.lineplot(x="percent_bin", y="propConstr",
             #hue="region", style="event",
             data=df)
```

```python
temp = df[['percent_bin', 'propTopical', 'propNonTopical']]
sns.lineplot(x='percent_bin', y='value', hue='variable', 
             data=pd.melt(temp, ['percent_bin']))
```

```python
temp = df[['percent_bin', 'propConstr', 'propRest']]
sns.lineplot(x='percent_bin', y='value', hue='variable', 
             data=pd.melt(temp, ['percent_bin']))
```

```python
sns.barplot(x='percent_bin', y='propNonTopical', color='Blue', 
             data=df)
sns.barplot(x='percent_bin', y='propTopical', color='Cyan', 
             data=df)
plt.show()
```

```python
temp = df.groupby(['percent_bin']).mean().reset_index()
```

```python
sns.set_palette("Paired")
# Data     propNonTopical     propTopical
x=temp.percent_bin
y=[ temp.propNonTopical, temp.propTopical ]
 
# Plot
plt.stackplot(x,y, labels=['propNonTopical','propTopical'])
plt.legend(loc='lower right')
plt.show()
```

```python
# Create data
x = temp.percent_bin
y1 = temp.propNonTopical
y2 = temp.propTopical

# Area plot
plt.fill_between(x, y1,color="skyblue", label='propNonTopical')
plt.fill_between(x, y2,color="slateblue", label='propTopical')
plt.legend(loc='right')
plt.show()
```

```python
# Data   propConstr   propRest
x=temp.percent_bin
ones = [1] * len(x)
y=[ temp.propRest, temp.propConstr]
 
# Plot
plt.stackplot(x,y, labels=['propRest','propConstr'])
plt.legend(loc='lower right')
plt.show()
```

```python
# Data   propConstr   propRest
x = temp.percent_bin
ones = [1] * len(x)
y = [temp.propRest, temp.propTopical, temp.propNonTopical]

# Plot
ax = plt.stackplot(x,y, labels=['Rest', 'Referential', 'Non-referential'], colors=['#3da4ab', '#f6cd61', '#fe8a71'])  #sns.color_palette('pastel')[3:])
plt.legend(loc='lower right')
plt.ylim(ymin=0.6)
# plt.show()
plt.savefig('/Users/mario/code/exp-rep/paper/aacl2022/submission/figures/proportions.pdf')
```
