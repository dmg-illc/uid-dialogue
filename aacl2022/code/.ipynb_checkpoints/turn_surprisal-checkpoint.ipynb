```python
df = pd.read_csv(
    '../estimates/surprisal_SBNC_gpt2_50_1e-3_agg.csv'
)
# df = pd.read_csv(
#     '../estimates/surprisal_SBNC_gpt2_50_0e+0.csv'
# )
```

```python
turn_s = []
for i, row in df.iterrows():
    token_surprisal = eval(row['Surprisal'])
    avg_s = np.mean(token_surprisal)
    if avg_s:
        turn_s.append(avg_s)
    else:
        raise ValueError()

df['TurnSurprisal'] = turn_s
```

```python
dlens = defaultdict(int)
for _, row in df.iterrows():
    dlens[row['Dialogue ID']] += 1
```

```python
idprop = []
for _, row in df.iterrows():
    idprop.append((row['Turn index'] + 1) / dlens[row['Dialogue ID']])
df['Turn index prop'] = idprop
```

```python
idprop = []
for _, row in chains.iterrows():
    idprop.append((row['CurrentTurn'] + 1) / dlens[row['Dialogue ID']])
chains['CurrentTurnProp'] = idprop
```

```python
bins = []
for _, row in chains.iterrows():
    bins.append(round(row['CurrentTurnProp'] * 100))
chains['Bin'] = bins
```
